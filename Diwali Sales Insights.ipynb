```python
#Count based on Gender
ax=s.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Sales based on Gender
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
s.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
#Which age group and gender contributes most to the sales
ax=s.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Age Group VS Amount
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
s.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
#Total no. of orders from top 10 states
sales_states=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
s.set(rc={'figure.figsize':(15,5)})
s.barplot(x='State',y='Orders',data=sales_states)
```

```python
#Total Amount VS sales from Top 10 states
sales_states=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
s.set(rc={'figure.figsize':(15,5)})
s.barplot(x='State',y='Amount',data=sales_states)
```

```python
#Count based on Marital Status
ax=s.countplot(x='Marital_Status',data=df)
s.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Which Gender whether married or unmarried contributes most to the sales
sales_ms=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
s.set(rc={'figure.figsize':(7,5)})
s.barplot(x='Marital_Status',y='Amount',data=sales_ms,hue='Gender')
```

```python
#Count based on Occupation sector
ax=s.countplot(x='Occupation',data=df)
s.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Occupation VS total sales
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
s.set(rc={'figure.figsize':(20,7)})
s.barplot(x='Occupation',y='Amount',data=sales_occupation)
```

```python
# the product_category dominates the most
ax=s.countplot(x='Product_Category',data=df)
s.set(rc={'figure.figsize':(30,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Product_Category VS Total Amount
sales_pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
s.set(rc={'figure.figsize':(35,10)})
s.barplot(x='Product_Category',y='Amount',data=sales_pc)
```

```python
#Top 10 sold products
sales_pid=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
s.set(rc={'figure.figsize':(20,7)})
s.barplot(x='Product_ID',y='Orders',data=sales_pid)
```

```python
#top 10 sold products (same as the above)
fig1, ax1=plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
